```python
#Imports
import xarray as xr
import matplotlib.pyplot as plt
```

```python
datashare = '/glade/p/datashare/marsh/waccm_cmip6/'
```

```python
hist = xr.open_dataset(datashare+\
    'historical-WACCM.001/b.e21.BWHIST.f09_g17.CMIP6-historical-WACCM.001.cam.h0.O3.2006-01-01_2014-12-01.nc')
ssp126 = xr.open_dataset(datashare+\
    '/SSP1-2.6-WACCM.001/b.e21.BWSSP126cmip6.f09_g17.CMIP6-SSP1-2.6-WACCM.001.cam.h0.O3.2046-01-01_2055-01-01.nc')
ssp245 = xr.open_dataset(datashare+\
    'SSP2-4.5-WACCM.001/b.e21.BWSSP245cmip6.f09_g17.CMIP6-SSP2-4.5-WACCM.001.cam.h0.O3.2046-01-01_2055-01-01.nc')
ssp370 = xr.open_dataset(datashare+\
    'SSP3-7.0-WACCM.001/b.e21.BWSSP370cmip6.f09_g17.CMIP6-SSP3-7.0-WACCM.001.cam.h0.O3.2046-01-01_2055-01-01.nc')
```

```python
#surface vmrs in ppb
hist_O3 = hist.O3.isel(lev=69).mean(dim='time')*1e9
ssp126_O3  = ssp126.O3.isel(lev=69).mean(dim='time')*1e9
ssp245_O3  = ssp245.O3.isel(lev=69).mean(dim='time')*1e9
ssp370_O3  = ssp370.O3.isel(lev=69).mean(dim='time')*1e9
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10,8))
fig.suptitle('Surface Ozone VMR (ppb)')

hist_O3.plot(ax =axs[0,0])
axs[0,0].set_title('Present Day')
axs[0,0].set_xlabel('')

diff_min = -15
diff_max = 15

(ssp126_O3-hist_O3).plot(ax =axs[0,1], vmin=diff_min, vmax = diff_max, cmap='RdBu_r')
axs[0,1].set_title('SSP1-2.6 - Present Day')
axs[0,1].set_xlabel('')
axs[0,1].set_ylabel('')


(ssp245_O3-hist_O3).plot(ax =axs[1,0], vmin=diff_min, vmax = diff_max, cmap='RdBu_r')
axs[1,0].set_title('SSP2-4.5 - Present Day')

(ssp370_O3-hist_O3).plot(ax =axs[1,1], vmin=diff_min, vmax = diff_max, cmap='RdBu_r')
axs[1,1].set_title('SSP3-7.0 - Present Day')
axs[1,1].set_ylabel('')

plt.show()
```
