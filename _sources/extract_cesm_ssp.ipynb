```python
datashare = '/glade/p/datashare/marsh/waccm_cmip6/'

lat_min = 33.
lat_max = 62.
lon_min = -12.
lon_max = 21.

#out_start = "2046-01-01"
#out_stop =  "2055-01-01"
out_start = "2006-01-01"
out_stop =  "2014-12-01"
```

```python
def extract_from_tseries(da, outfile):

    # shift dataset from 0 to 360 lon to -180 to 180
    da = da.roll(lon=144, roll_coords=True)
    lon = da['lon']
    da['lon'] = xr.where(lon >= 180, lon-360, lon)
    
    # subset the dataset to region of interest 
    # write to datashare

    da.sel(lat=slice(lat_min, lat_max), \
           lon=slice(lon_min, lon_max)).to_netcdf(outfile)
    
    return
```

```python
def process_case(casename,outdir):
    
    species_list = '../notes/waccm_species_to_select.txt'

    file1 = open(species_list, 'r')

    for species in file1:

        field = species.strip()
        ncfile = archive_dir+casename+monthly+casename+'.cam.h0.'+field+date_range
        print(field)
        ds = xr.open_dataset(ncfile)
        da = ds[field].sel(time=slice(out_start, out_stop))

        out_file = datashare+outdir+casename+'.cam.h0.'+field+'.'+out_start+'_'+out_stop+'.nc'

        print(out_file)
        extract_from_tseries(da, out_file)
        ds.close()

    file1.close()
    
    return
```
